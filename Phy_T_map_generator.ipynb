```python
import numpy as np
import pandas as pd

data = [[2006.82798,1762.00623,1190.038933,1543.693609,1897.768821,2125.158297,1930.692597,1509.12494,1221.369788
],[2.276264981,2.833434029,3.498807663,3.395279403,2.79160087,2.252623014,2.301624621,2.865139007,3.882156967
], [9.568,8.69,0.211,3.353,9.974,9.602,9.653,3.184,0.552]]
array = np.array(data).T

index1 = [i for i in range(1, 10)]
column1 = ["Temperature", "Equivalence ratio", "smoke"]

df = pd.DataFrame(data=array, index=index1, columns=column1)
df
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.size']=16

x_name = 'Temperature'
y_name = 'Equivalence ratio'
z_name = 'smoke'

fig, ax = plt.subplots()

sc = ax.scatter(df[x_name], df[y_name],
                c = df[z_name],
                cmap = 'bwr')

for i, txt in enumerate(df.index.values):
    plt.annotate(txt, (df[x_name].values[i], df[y_name].values[i]))

fig.colorbar(sc)
ax.set_xlabel(x_name)
ax.set_ylabel(y_name)
ax.grid()
ax.set_xlim(1000, 2500)
ax.set_ylim(1, 4)
plt.title(z_name)
plt.tight_layout()
plt.savefig("/content/drive/MyDrive/phy_T_map20230831.png")
plt.show()
```
