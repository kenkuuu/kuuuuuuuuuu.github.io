```python
import numpy as np
import pandas as pd

data = [[1537.873,1519.805,1523.871,1542.064,1550.811,1579.143,1501.855,1484.627,1513.134
],[2.276265,2.833434,3.498808,3.395279,2.791601,2.252623,2.301625,2.865139,3.882157], [9.568,8.69,0.211,3.353,9.974,9.602,9.653,3.184,0.552]]
array = np.array(data).T

index1 = [i for i in range(1, 10)]
column1 = ["Temperature", "Equivalence ratio", "smoke"]

df = pd.DataFrame(data=array, index=index1, columns=column1)
df
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.size']=16

x_name = 'Temperature'
y_name = 'Equivalence ratio'
z_name = 'smoke'

fig, ax = plt.subplots()

sc = ax.scatter(df[x_name], df[y_name],
                c = df[z_name],
                cmap = 'bwr')

fig.colorbar(sc)
ax.set_xlabel(x_name)
ax.set_ylabel(y_name)
ax.grid()
plt.title(z_name)
plt.tight_layout()
plt.show()
```
